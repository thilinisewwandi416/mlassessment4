```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import joblib  
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/BostonHousing.csv')

y = df['medv']
x = df.drop('medv', axis=1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)

lr = LinearRegression()
lr.fit(x_train, y_train)

joblib.dump(lr, 'linear_regression_model.pkl')

loaded_model = joblib.load('linear_regression_model.pkl')

y_lr_train_pred = loaded_model.predict(x_train)
y_lr_test_pred = loaded_model.predict(x_test)

print("Training Set Predictions:")
print(y_lr_train_pred)
print("\nTesting Set Predictions:")
print(y_lr_test_pred)

plt.scatter(x=y_train, y=y_lr_train_pred, alpha=0.3)
plt.title('Training Set: Actual vs Predicted')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()
```
